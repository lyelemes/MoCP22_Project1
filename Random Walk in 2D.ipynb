```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) 
def RandomWalk(N):
    walk_x = [0]
    walk_y = [0]
    
    x = [0, 0]
    
    for i in range(N):
        delta_x_prime = np.random.uniform(-1, 1, 2) 
        delta_x = delta_x_prime/np.linalg.norm(delta_x_prime)
        x = x + delta_x
        walk_x.append(x[0])
        walk_y.append(x[1])
    
    return walk_x, walk_y
```

```python
N=1000


x, y = RandomWalk(N)
plt.plot(x, y, color = 'r')
plt.title('2D-Random Walk (N = 1000)')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('Random Walk 1.png')
plt.show()

x, y = RandomWalk(N)
plt.plot(x, y, color = 'b')
plt.title('2D-Random Walk (N = 1000)')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('Random Walk 2.png')
plt.show()

x, y = RandomWalk(N)
plt.plot(x, y, color = 'g')
plt.title('2D-Random Walk (N = 1000)')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('Random Walk 3.png')
plt.show()
```

```python
# b) M = 1000 N = 10000

def R_N(M, N):
    r = []
    
    for i in range(M):
        x, y = RandomWalk(N)

        x_end = x[-1]
        y_end = y[-1]

        r.append(np.sqrt(x_end**2 + y_end**2))
    
    return r
```

```python
M = 1000
N = 10000


plt.hist(R_N(M, N), 40)
plt.title('Histogram of $R_N$ after 1000 simulations (N = 10000)')
plt.xlabel('$R_N$')
plt.ylabel('Frequency')
plt.savefig('Random Walk hist.png')
plt.show()
```

```python
def R_rms(N):
    delta_r = 0
    
    for i in range(N):
        delta_x_prime = np.random.uniform(-1, 1, 2) 
        delta_x = delta_x_prime/np.linalg.norm(delta_x_prime)
        
        delta_r = delta_x[0]**2 + delta_x[1]**2 + delta_r
        
    return np.sqrt(delta_r)
```
